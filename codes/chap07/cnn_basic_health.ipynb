```python
hd = HealthDataset([96, 96], [96, 96, 3])
```

```python
hm1 = CnnBasicModel('health_model_1', hd, [30, 10])
hm1.exec_all(epoch_count = 1, report = 1,show_cnt=3)
```

```python
hd.init_visualize_data([96, 96], [96, 96, 3])
hm1.visualize(1)
```

```python
hm2 = CnnBasicModel('health_model_2', hd, 
                    [['full', {'width':30}], 
                     ['full', {'width':10}]])
hm2.use_adam = False
hm2.exec_all(epoch_count = 1, report = 1)
```

```python
hm3 = CnnBasicModel('health_model_3', hd, 
               [['conv', {'ksize':5, 'chn':6}], 
                ['max', {'stride':4}], 
                ['conv', {'ksize':3, 'chn':12}], 
                ['avg', {'stride':2}]], 
               True)
hm3.exec_all(epoch_count = 10, report = 2)
```

```python
hm4 = CnnBasicModel('health_model_4', hd, 
               [['conv', {'ksize':3, 'chn':6}], 
                ['max', {'stride':2}], 
                ['conv', {'ksize':3, 'chn':12}], 
                ['max', {'stride':2}], 
                ['conv', {'ksize':3, 'chn':24}], 
                ['avg', {'stride':3}]])
hm4.exec_all(epoch_count = 100, report = 20)
```

```python
hm5 = CnnBasicModel('health_model_5', hd, 
               [['conv', {'ksize':3, 'chn':6}], 
                ['max', {'stride':2}], 
                ['conv', {'ksize':3, 'chn':12}], 
                ['max', {'stride':2}], 
                ['conv', {'ksize':3, 'chn':24}], 
                ['max', {'stride':2}], 
                ['conv', {'ksize':3, 'chn':48}], 
                ['max', {'stride':2}], 
                ['conv', {'ksize':3, 'chn':96}], 
                ['avg', {'stride':3}]])
hm5.exec_all(epoch_count = 100, report = 20)
```

```python
hm6 = CnnBasicModel('health_model_6', hd, 
               [['conv', {'ksize':3, 'chn':6}], 
                ['max', {'stride':2}], 
                ['conv', {'ksize':3, 'chn':12}], 
                ['max', {'stride':2}], 
                ['conv', {'ksize':3, 'chn':24}], 
                ['max', {'stride':2}], 
                ['conv', {'ksize':3, 'chn':48}], 
                ['avg', {'stride':3}]])
hm6.exec_all(epoch_count = 100, report = 20)
```
